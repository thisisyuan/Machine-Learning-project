# Part E: Train-Test Split

**E1)** 
Create a training-test split function called “My_split”. The function shuffles the input data and selects random samples to generate training and test data sets which is determined by another user input “ratio” between [0.0-1.0]. 

Divide the data frame of “personAclick” into training and test sets with a ratio of 0.8 (80% Training, 20% Test). What are the dimensions of the train and test data frames? Why do we need this process for machine learning applications?


```python
def My_split(data,ratio):
    #random seed
    np.random.seed(1)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)) * ratio
    test_indices = shuffled_indices[:int(test_set_size)]
    train_indices = shuffled_indices[int(test_set_size):]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
A = np.load('personAclick.npy')
z = pd.DataFrame(A)
train_set, test_set = My_split(z,0.2)
```

The purpose of training set is to train a model using machine learning algorithm，and testing set is used to verify  and evaluate the trained model then see if the model works. No training should be performed on testing set!
